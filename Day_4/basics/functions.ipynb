```python
# create a function to add/subtract two numbers 
def add(a,b):
    return a+b
print(add(3,4)) #7

def sub(a,b):
    return a-b
print(sub(3,4)) #-1

# add three numbers 
def add(a,b,c):
    return a+b+c

print(add(1,2)) #error
print(add(1,2,3))



```

```python
#for n number of inputs
def add_n(*args):
    print(args)
    op=0
    for i in args:
        op+=i
    return op

print(add_n(1,2,3,4))
print(add_n(1,2,3,4,5,6))
```

```python
# global keyword 
x=10
def fun(y):
    global x
    x+=y
    return x

print(x)
fun(20) 
fun(25)
print(x)
```

```python
# lambda function 
add=lambda a,b: a+b

print(add(3,4))

make_capital=lambda s: s.upper()

print(make_capital("hello world"))
```

```python
# map functions 
arr=[1,2,3,4]
square=lambda a: a*a

square_array=map(square,arr)
print(square(4))
print(square_array)
print(list(square_array))
```

```python
#filter

is_even=lambda x: x%2==0
# Filter even numbers
nums=list(range(1,21))
evens=list(filter(is_even,nums))
print(f"Even from 1-20:{evens}")
```

```python
from functools import reduce 

#sum of numbers 
nums=[1,2,3,4,5]
total=reduce(lambda a,b: a+b, nums)
print(total)
```

```python
#recursion 

def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)

print(factorial(5))
```

```python
# factorial using for-loop 

def factorial(n):
    op=1
    for i in range(1,n+1):
        op*=i
    return op

print(factorial(5))

```

```python
# generator funcrion 
def count_up_to(n):
    count=1
    print("Before while loop")
    while count<=n:
        yield count 
        print(f"Yielded: {count}")
        count+=1
    print("done")

#count_up_to(5)
for num in count_up_to(5):
    print(num)

# examples of inbuilt generators 
# range 
# enumerate 
# zip
"""generators are used in production like in data processing, web scraping etc, 
where we need to process large data sets without loading everything to memeory
 at once"""
```
